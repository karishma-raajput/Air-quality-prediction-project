```python
# converting th date from DD/MM/YYYY to YYYY-MM-DD
date_info=pd.to_datetime(air_data['Date'])
print(date_info)
```

```python
time_info=air_data['Time']
print(time_info)

```

```python
time_info =time_info.apply(lambda x:x.replace('.',':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# combining 2series to a pandas dataframe
date_time=pd.concat([date_info,time_info],axis=1)
```

```python
# combining date and time
date_time['ds']=date_time['Date'].astype(str)+' ' + date_time['Time'].astype(str)
```

```python
from prophet import Prophet
```

```python
# 1 training the prophet model
model=Prophet()
model.fit(data)
```

```python
# for future prediction

future =model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
# ds is date and time information
# yhat is kind of middlevalue 

forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_lower']].tail()
```
